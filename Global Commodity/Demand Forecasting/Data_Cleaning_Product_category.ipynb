## Generate the product category data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('New_file.csv')
df
```

```python
del df['Unnamed: 0']
del df['Unnamed: 0.1']
df
```

```python
df_import = df[df['FLOW']=='import']
df_import
```

```python
df_import = df_import.reset_index()
del df_import['index']
df_import
```

```python
df_import_clean = df_import[df_import['PRODUCT_ID'].str.contains('^\d+$')]
```

```python
df_import_clean
```

```python
df_import_clean = df_import_clean.reset_index()
del df_import_clean['index']
df_import_clean
```

```python
df_import_clean_sorted = df_import_clean.sort_values('PRODUCT_ID')
df_import_clean_sorted
```

```python
df_import_clean_sorted_idx = df_import_clean_sorted.reset_index()
del df_import_clean_sorted_idx['index']
df_import_clean_sorted_idx
```

```python
for i in range(len(df_import_clean_sorted_idx['PRODUCT_ID'])):
  if(int(int(df_import_clean_sorted_idx['PRODUCT_ID'][i])/1000000)==62):
    break

start_idx = i
```

```python
for i in range(len(df_import_clean_sorted_idx['PRODUCT_ID'])):
  if(int(int(df_import_clean_sorted_idx['PRODUCT_ID'][i])/1000000)==63):
    break

end_idx = i
```

```python
df_import_clean_sorted_idx['PRODUCT_ID'][start_idx]
```

```python
df_import_clean_sorted_idx['PRODUCT_ID'][end_idx]
```

```python
df_cat = df_import_clean_sorted_idx.iloc[start_idx:end_idx]
```

```python
df_cat
```

```python
df_cat_idx = df_cat.reset_index()
del df_cat_idx['index']
df_cat_idx
```

```python
df_cat_quantity = df_cat_idx.groupby(['PERIOD']).QUANTITY_IN_KG.sum()
df_cat_quantity
```

```python
quantity = []
for i in range(len(df_cat_quantity)):
  quantity.append(df_cat_quantity.iloc[i])

quantity
```
